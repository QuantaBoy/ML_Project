Importing Important Libraries

Reading The File Using Pandas

```python
import pandas as pd
pharma_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/pharma_data/Training_set_begs.csv')
```

Preview Of The Data

```python
pharma_data.head()
```

To Find The Number Of Columns and Rows

```python
pharma_data.shape
```

Dropping The Independent Variable

```python
inputs = pharma_data.drop(['ID_Patient_Care_Situation','Patient_ID','A','B','C','D','E','F','Z','Patient_mental_condition'],axis ='columns')
```

```python
inputs
```

Converting The Status Data Into Binary Data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_drugs = LabelEncoder()
le_smoker = LabelEncoder()
le_Place = LabelEncoder()
```

```python
inputs['drugs_n'] = le_drugs.fit_transform(inputs['Treated_with_drugs'])
inputs['smoker_n'] = le_smoker.fit_transform(inputs['Patient_Smoker'])
inputs['Place_n'] = le_Place.fit_transform(inputs['Patient_Rural_Urban'])
```

```python
pharma_data.dropna()
```

```python
pharma_data.info()
```

```python
inputs
```

Dropping The Datasets From The Input

```python
pharma_data['Number_of_prev_cond'].max()
```

```python
inputs = inputs.drop(['Treated_with_drugs','Patient_Smoker','Patient_Rural_Urban','Survived_1_year'],axis='columns')
```

```python
inputs.to_csv("Cleaned Dataset.csv")
```

Taking The Target Variable From The Datasets(Survived_1_year)

```python
target = pharma_data['Survived_1_year']
```

```python
target
```

Importing Important Libraries

```python
from sklearn import tree
```

Using The DesicionTreeClassifier

```python
model = tree.DecisionTreeClassifier()
```

Fitting The Model

```python
model.fit(inputs,target)
```

Model Accuracy Rate

```python
Accuracy = model.score(inputs, target)
Accuracy = Accuracy*100
print(f"Model Accuracy : {Accuracy:.2f} %")
```

-------MODEL TRAINING COMPLETED-------

```python
import pickle
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)
```
