Importing Important Libraries

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

Reading The Datasets Using Pandas

```python
df = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/startupdata/training_set_label.csv')
```

Preview Of Datasets

```python
df.head()
```

To Find The Number Of Columns and Rows

```python
df.shape
```

Dropping The Independent Variable

```python
inputs = df.drop(['Unnamed: 0','state_code','latitude','longitude','zip_code','id','city','Unnamed: 6','name','object_id'
                    ,'founded_at','closed_at','first_funding_at','last_funding_at','age_first_milestone_year','age_last_milestone_year','funding_rounds'
                    ,'state_code.1','object_id','category_code','is_CA','is_NY','is_MA','is_TX','is_otherstate','has_VC','has_angel','has_roundA','has_roundB',
                    'has_roundC','has_roundD','funding_rounds','funding_total_usd','age_first_funding_year','avg_participants'],axis = 'columns')
```

```python
inputs
```

Converting The Status Data Into Binary Data

```python
le_status = LabelEncoder()
```

```python
inputs['status_n'] = le_status.fit_transform(inputs['status']) #reason for converting as binary : The Model will understand the numerical values so we converting it
```

Taking The Target Variable From The Datasets(Status)

```python
target = inputs['status_n']
```

```python
target
```

Once The Target Variable Is Took as Target we Have To Drop It From The Input(Reason:Target Variable Is the Data We Have To find)

```python
inputs = inputs.drop(['status_n','status'],axis = 'columns')
```

```python
inputs
```

Using The LogisticRegression Algorithm

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

Splitting The Data Into Two Parts Namely,Train & Test(X,Y)
Inputs as X
Target as Y

```python
from sklearn.model_selection import train_test_split 
X_Train,X_Test,Y_Train,Y_Test = train_test_split(inputs,target,train_size = 0.7,random_state = 42)
```

Fit The Model With The Train Datasets

```python
model.fit(X_Train,Y_Train)
```

Predict The Model Using The Training Data

```python
y_pred = model.predict(X_Test)
```

In Confusion Matrix 

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_Test, y_pred) 
#1.TP Stands For True Positive 
# 2.TN Stands For True Negative 
# 3.FN Stands For False Negative 
# 4.FP Stands For False Positive
```

```python
TP,FP,TN,FN = cm.ravel()
```

```python
print(f"True Positive {TP}")
print(f"False Positive {FP}")
print(f"True Negative {TN}")
print(f"False Negative {FP}")
```

Accuracy Formula

```python
accuracy = (TP+TN)/TP+FP+FN+TN
accuracy
```

Finding The Accuracy of The Model Using Inbuilt Accuracy Keyword

```python
from sklearn.metrics import accuracy_score
```

```python
Accuracy = model.score(X_Test, Y_Test)
Accuracy = Accuracy*100
print(f"Model Accuracy : {Accuracy:.2f} %")
```
